# **การทำให้ข้อมูลเป็นมาตรฐาน เพราะบางครั้งเราอาจพิมว่าวิทย์แต่เราหมายถึงคณะวิทยาศาสตร์**

```python
def standardize_faculty(name):
    if pd.isna(name):
        return name  # ถ้าเป็น NaN ก็คืนค่า NaN ไปเลย

    name = str(name).strip()  # ตัดช่องว่างหน้า-หลัง

    # กลุ่มวิทยาศาสตร์
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'Sci', 'คณะวิทยาศาสตร์']:
        return 'วิทยาศาสตร์'

    # กลุ่มศิลปศาสตร์
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'

    # กลุ่มวิศวกรรมศาสตร์
    elif name in ['วิศวะ', 'วิศวกรรม', 'วิศวกรรมศาสตร์', 'Engineering','วิศวฯ']:
        return 'วิศวกรรมศาสตร์'

    # อื่น ๆ ให้คืนค่าตามเดิม
    else:
        return name

# ใช้งานกับ DataFrame
data['คณะ'] = data['คณะ'].apply(standardize_faculty)
```

# ทำให้เพศเป็นมาตราฐานเดียวกันด้วย

```python
def standardize_gender(g):
    if pd.isna(g):
        return g  # ถ้าเป็น NaN ก็คืน NaN

    g = str(g).strip().lower()  # ตัดช่องว่างและแปลงเป็นตัวพิมพ์เล็ก

    # กลุ่มชาย
    if g in ['ชาย', 'ช', 'male', 'm']:
        return 'ชาย'

    # กลุ่มหญิง
    elif g in ['หญิง', 'ญ', 'female', 'f']:
        return 'หญิง'

    # อื่น ๆ คืนตามเดิม (กันกรณีมีเพศพิเศษ)
    else:
        return g

# ใช้กับ DataFrame
data['เพศ'] = data['เพศ'].apply(standardize_gender)
```

# วิธีการจัดการค่าผิดปกติ ตัวอย่างเช่น  GPA: -1.00 หรือ อายุ150

# จัดการกับอายุที่ผิดปกติ

```python
#ใช้ IQR (Interquartile Range)
Q1 = data['อายุ'].quantile(0.25)
Q3 = data['อายุ'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# กรองค่าผิดปกติ
df_no_outliers = data[(data['อายุ'] >= lower_bound) & (data['อายุ'] <= upper_bound)]
```

```python
df_no_outliers
```

# จัดการกับGPAที่ติดลบเช่น -1.00 โดยให้มันกลายเป็นค่าเฉลี่ย


```python
mean_gpa = data['GPA'].mean()
data['GPA'] = data['GPA'].replace(-1.00, mean_gpa)
data['GPA'] = data['GPA'].round(2)
data
```

**จัดการกับGPAที่ติดลบเช่น GPAมากกว่า4 โดยการตั้งเงื่อนไข** เพื่อไม่ให้เกิดข้อมูลที่ผิดพลาด

```python
mean_gpa = data['GPA'].mean()
data.loc[ (data['GPA'] > 4.0), 'GPA'] = mean_gpa
data
```

แก้ไขอายุที่เป็นค่าผิดปกติ (เช่น 150.0) ให้เป็นฐานนิยม

```python
mode_age = data['อายุ'].mode()[0]
data['อายุ'] = data['อายุ'].replace({150.0: mode_age})
data['GPA'] = data['GPA'].round(2)
data
```
